```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score

# encoding categorical variables
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['item_type']])
X_ohe = ohe.fit_transform(X[['item_type']]).toarray()
ohe2 = OneHotEncoder(handle_unknown='ignore')
ohe2.fit(X[['status']])
X_be = ohe2.fit_transform(X[['status']]).toarray()
# independent features after encoding
X = np.concatenate((X[['quantity_tons', 'application', 'thickness', 'width','country','customer','product_ref']].values, X_ohe, X_be), axis=1)
scaler = StandardScaler()
X = scaler.fit_transform(X)
# test and train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
#Linear Regression

from sklearn.linear_model import LinearRegression


# Create and fit a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the R-squared (R2) score
r2 = r2_score(y_test, y_pred)
print('R-squared (R2) Score:',r2)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt



# Define a range of K-values to test
k_values = list(range(1, 21))  # You can adjust the range as needed

# Initialize lists to store the cross-validation results
cv_scores = []

# Perform cross-validation for each K-value
for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn_regressor, X_train, y_train, cv=5, scoring='r2')  # Use 'r2' scoring
    cv_scores.append(np.mean(scores))

# Find the optimal K-value that maximizes the R2 score
optimal_k = k_values[cv_scores.index(max(cv_scores))]

print(f'Optimal K-value: {optimal_k}')

# Visualize the cross-validation results (optional)
plt.plot(k_values, cv_scores, marker='o', linestyle='-')
plt.title('Cross-Validation Results for Different K-Values')
plt.xlabel('K-Value')
plt.ylabel('R-squared (R2) Score')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

def tune_decision_tree(X_train, y_train, X_test, y_test, random_state=None):
    # Define the hyperparameter grid
    param_grid = {
        'max_depth': [2, 5, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['auto', 'sqrt', 'log2']
    }

    # Create a Decision Tree Regressor with random_state
    dtr = DecisionTreeRegressor(random_state=random_state)

    # Create a GridSearchCV instance
    grid_search = GridSearchCV(estimator=dtr, param_grid=param_grid, cv=5)

    # Fit the model
    grid_search.fit(X_train, y_train)

    # Get the best hyperparameters
    best_params = grid_search.best_params_
    print("Best hyperparameters:", best_params)

    # Get the best model
    best_model = grid_search.best_estimator_

    # Make predictions
    y_pred = best_model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print('Mean squared error:', mse)
    print('R-squared:', r2)

# Usage
tune_decision_tree(X_train, y_train, X_test, y_test, random_state=42)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

def tune_random_forest(X_train, y_train, X_test, y_test, random_state=None):
    # Define the hyperparameter grid
    param_grid = {
        'max_depth': [10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['auto', 'sqrt', 'log2']
    }

    # Create a Random Forest Regressor with a fixed number of estimators (100) and random_state
    rfr = RandomForestRegressor(n_estimators=100, random_state=random_state)

    # Create a GridSearchCV instance
    grid_search = GridSearchCV(estimator=rfr, param_grid=param_grid, cv=5)

    # Fit the model
    grid_search.fit(X_train, y_train)

    # Get the best hyperparameters
    best_params = grid_search.best_params_
    print("Best hyperparameters:", best_params)

    # Get the best model
    best_model = grid_search.best_estimator_

    # Make predictions
    y_pred = best_model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print('Mean squared error:', mse)
    print('R-squared:', r2)

# Usage
tune_random_forest(X_train, y_train, X_test, y_test, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, f1_score
from collections import Counter
from sklearn.linear_model import LogisticRegression



# Define a list of potential sampling strategies to test
sampling_strategies = [0.2, 0.4, 0.6, 0.8, 'auto']

# Define a scoring function based on the metric you want to optimize (e.g., F1-score)
scorer = make_scorer(f1_score)

# Initialize an empty dictionary to store results
results = {}

# Loop through the sampling strategies
for strategy in sampling_strategies:
    if strategy == 'auto':
        sampler = RandomOverSampler(sampling_strategy=strategy, random_state=42)
    else:
        sampler = SMOTE(sampling_strategy=strategy, random_state=42)
    
    # Perform cross-validation with the current sampling strategy
    scores = cross_val_score(LogisticRegression(random_state=42), X, y, cv=5, scoring=scorer, fit_params={'sampling_strategy': sampler})
    
    # Calculate the mean F1-score across folds
    mean_f1_score = scores.mean()
    
    # Store the results in the dictionary
    results[str(strategy)] = mean_f1_score

# Find the best sampling strategy based on F1-score
best_strategy = max(results, key=results.get)
best_f1_score = results[best_strategy]

print("Results for each sampling strategy:")
for strategy, f1_score in results.items():
    print(f"Sampling Strategy: {strategy}, Mean F1-score: {f1_score}")

print(f"Best Sampling Strategy: {best_strategy}, Best Mean F1-score: {best_f1_score}")
```
